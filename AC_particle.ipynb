#particle pair correlation $g(r)$

###1.pick a value of $dr$. this will be the resolution

###2.compute the particles' distance $R$ for every time step you want.Please take care of the periodic boundary condition.As a result, each pair has each distance. Let me call this distance as "bond of particles' pair". 
   
###3.count the number of the partcles' pairs  such that the particle distances are among $r$ to $r+dr$. Or, equally, count the bond such that  $r$ to $r+dr$.and store the data separated by each  $r$  value.Say this counts are $G(r)$.
   
###4.Devide this $G(r)$ by $4\pi r^{2}\cdot n(r)$ where $n(r)$ as the number density.Again notice that PBC when you considering $n(r)$. The result is the pair correlation function $g(r)$. 
 
ここ、もう一度データを解析しなおし
rmax=32までで、30点とってプロットした方がいいだろう

```python
file_pc05 = AC+"rad_paircorr_5.txt"
file_pc3 = AC+"rad_paircorr_30.txt"

R_g05 = np.loadtxt(file_pc05)
R_g3 = np.loadtxt(file_pc3)


def gr_plot(fig,ax,gr,cycle,interval,flg):

    for i in np.arange(1,cycle,interval):
        if flg:
            c = cm.autumn(i/cycle,11)
           
        elif flg==False:
            c = cm.jet(i/cycle,11)
            
        alpha = 0.1+0.9*(i/cycle)
        ax.plot(gr[:,0],
                gr[:,i],
                color=c,
                alpha=alpha,
                                )
 
iv=1

fig_gr,[ax_gr3,ax_gr05]=plt.subplots(2,1,sharex=True,figsize=(13,10))

ax_gr05.set_xlabel('$r/2a$',fontsize=labelsize)
ax_gr05.set_yscale('log')
ax_gr3.set_yscale('log')
#ax_gr05.set_xlim(1,4)
ax_gr05.set_ylabel('$g(r)$',fontsize=labelsize)
ax_gr3.set_ylabel('$g(r)$',fontsize=labelsize)

gr_plot(fig_gr,ax_gr05,R_g05,cycle_g05,iv,0)
gr_plot(fig_gr,ax_gr3,R_g3,cycle_g3,iv,1)

fig_gr2,ax_gr=plt.subplots(figsize=(13,7.5))

ax_gr.set_xlabel('$r/2a$',fontsize=labelsize)
ax_gr.set_xlim(0,4)
ax_gr.set_ylim(10e-2,10e0)
ax_gr.set_yscale('log')
ax_gr.set_ylabel('$g(r)$',fontsize=labelsize)
ax_gr.tick_params(labelsize=ticksize)
gr_plot(fig_gr2,ax_gr,R_g05,cycle_g05,iv,0)
gr_plot(fig_gr2,ax_gr,R_g3,cycle_g3,iv,1)

#ax_gr.legend(fontsize=18,loc='upper right')

fig_gr2.savefig(AC+'gr.png')
fig_gr2.savefig(AC+'gr.pdf')
```
