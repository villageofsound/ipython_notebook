```python
#prefix
%matplotlib inline
from pyrecipes import *
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.axes_grid1 import make_axes_locatable
import brewer2mpl
import seaborn as sns
sns.set_style('white')
colors = sns.color_palette()
greys  = brewer2mpl.get_map('Greys', 'Sequential', 9)
```

```python
#set parameters
#please change the name of fp to "withfp" if you want to see the  
fp = "withfp"
fpp ="/"+fp
tag = "./1particle"
fname = tag + fpp+"/data.h5"
radius= 8.0
Lx = Ly = 64.0
srate = 0.001
ss = 2#particle snapshots
nump, numf = get_h5_pconf(fname)#nump particle number, numf number of frames
print nump
print numf
```

```python
#get coordinate data
h5out = h5py.File(fname,'r')
x = h5out["/coord_data/x"].value[:,:,0]
y = h5out["/coord_data/y"].value[:,:,0]

print x,y
```

```python
#get the particle & field data
f0 = h5out["/frame_0/field_data/u_x"].value[:,:,0]
#particle
ploc = '/frame_'+str(ss)+'/particle_data/'
pos = h5out[ploc+"pos"].value

#field
ux = np.zeros(f0.shape, f0.dtype)
uy = np.zeros(f0.shape, f0.dtype)
tauxy = np.zeros(f0.shape,f0.dtype)
phi = np.zeros(f0.shape,f0.dtype)
us = srate*(y-Ly/2)
print f0.shape
#time average the data
count = 0
for i in np.arange(numf/2,numf,1):
    loc = "/frame_"+str(i)+"/field_data/"
    ux = ux + h5out[loc+"u_x"].value[:,:,0]
    uy = uy + h5out[loc+"u_y"].value[:,:,0]
    tauxy = tauxy + h5out[loc+"tau_xy"].value[:,:,0]
    phi = phi + h5out[loc+"phi"].value[:,:,0]
    count = count + 1
    
ux = ux / float(count)
uy = uy / float(count)
tauxy = tauxy/float(count)
phi = phi/float(count)

uxx = ux-us #subtract mean shear flow
uu = np.dstack((uxx,uy))#make a vector
u2 = np.linalg.norm(uu,axis=2)#compute the norm of velocity field
```

```python
#define plot functions

gmap = brewer2mpl.get_map('Greys', 'Sequential',6)
bmap = brewer2mpl.get_map('RdBu', 'Diverging',11,reverse=True)
bmap2 = brewer2mpl.get_map('YlOrRd', 'Sequential',8)

def draw_particles(fig, ax, pos, radius):
     for i in range(len(pos)):
            circ = plt.Circle(pos[i,0:2]-[Lx/2,Ly/2],radius=radius,
                              edgecolor=greys.mpl_colors[8],
                              facecolor='none',lw=2.0)
            ax.add_patch(circ)

def draw_reverse_stress(fig, ax, pos, radius):
     scalar = np.sqrt(19./15.)
     for i in range(len(pos)):
            circ = plt.Circle(pos[i,0:2],radius=scalar*radius,
                              alpha=0.3,
                              edgecolor='red',#greys.mpl_colors[8],
                              facecolor='none',lw=2.0)
            ax.add_patch(circ)

        
def draw_stress(fig, ax, phi, sxy, vmin, vmax):
    im = ax.imshow(((1-phi)*2.*sxy).T,
                   origin="lower", 
                   interpolation='bicubic', 
                   cmap=bmap.get_mpl_colormap(),alpha=0.5,
                    extent = [-Lx/2,Lx/2-1,-Ly/2,Ly/2-1],
                   vmin=vmin,vmax=vmax
                    )
    
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.15)
    tick =np.arange(-0.00040,0.000401,0.0002)
    cbar= fig.colorbar(im, cax,ticks=tick)
    cbar.ax.tick_params(labelsize=16)
    
def draw_vel(fig, ax, x, y, phi, ux, uy, skipx, skipy, vscale,width):
    ax.quiver(x[::skipx,::skipy]-Lx/2, y[::skipx,::skipy]-Ly/2,
              (1.0-phi[::skipx,::skipy])*ux[::skipx,::skipy],
              (1.0-phi[::skipx,::skipy])*uy[::skipx,::skipy],
              units='xy',angles='uv',scale=vscale,width=width)
def draw_u2map(fig,ax,unorm):
    im=ax.imshow((1.0-phi)*unorm.T, 
                 origin='lower',
                 interpolation="bicubic",
                 cmap=bmap2.get_mpl_colormap(),
                 extent = [-Lx/2,Lx/2-1,-Ly/2,Ly/2-1],
                 alpha=0.6,vmin=0, vmax=0.002
             )
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.15)
    tick =np.arange(0.0,0.021,0.0005)
    cbar= fig.colorbar(im, cax,ticks=tick)
    cbar.ax.tick_params(labelsize=16)
    #cbar.formatter.set_scientific(True)
    #cbar.formatter.set_powerlimits((-2,0)) 
    #cbar.update_ticks()                   

def draw_streamline(fig,ax,x,y,phi,ux,uy):
    ax.streamplot(y[0]-Ly/2,x[:,0]-Lx/2,(1.0-phi.T)*ux.T,(1.0-phi.T)*uy.T,
                  density=[1.5,1.5],linewidth=1,arrowsize=1.2
                  )
```

```python
#plot
taurange = 0.0004
skip = 2
vscale = 0.1
width = 0.3
fig, ax = plt.subplots(figsize=(8,7))

ax.xaxis.set_major_locator(LinearLocator(3))
ax.yaxis.set_major_locator(LinearLocator(3))
fig2, ax2 = plt.subplots(figsize=(8,7))
ax2.xaxis.set_major_locator(LinearLocator(3))
ax2.yaxis.set_major_locator(LinearLocator(3))
#integrate 2 fig.
#fig3, [ax,ax2] =plt.subplots(1,2,sharey=True,figsize=(12,5)) 

ax.set_title("numerical",fontsize=18)

ax.set_xlabel('$x$',fontsize = 26)
ax.set_ylabel('$y$',fontsize = 26)
ax.tick_params(labelsize=22)


#ax.set_xticklabels(np.arange(0,64,10),fontsize=16)
#ax.set_yticklabels(np.arange(0,64,10),fontsize=16)
draw_particles(fig,ax,pos,radius)
draw_stress(fig,ax,phi,tauxy,-taurange,taurange)
draw_vel(fig,ax,x,y,phi,uxx,uy,skip,skip,vscale*srate*radius,width)


ax2.set_title("numerical",fontsize=18)
ax2.set_xlabel('$x$',fontsize = 26)
ax2.set_ylabel('$y$',fontsize = 26)
ax2.tick_params(labelsize=22)
#ax2.set_xticklabels(np.arange(0,64,10),fontsize=16)
#ax2.set_yticklabels(np.arange(0,64,10),fontsize=16)
draw_u2map(fig2,ax2,u2)
draw_streamline(fig2,ax2,x,y,phi,uxx,uy)
draw_particles(fig,ax2,pos,radius)

#ax.set_xlim(0,Lx)
#ax.set_ylim(0,Ly)
#ax2.set_xlim(0,Lx)
#ax2.set_ylim(0,Ly)
ax.set_xlim(-Lx/2,Lx/2)
ax.set_ylim(-Ly/2,Ly/2)
ax2.set_xlim(-Lx/2,Lx/2)
ax2.set_ylim(-Ly/2,Ly/2)
```
