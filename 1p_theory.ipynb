# 1 particle velocity/stress field(analytical ) 

```python
#prefix
%matplotlib inline
from pyrecipes import *
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import brewer2mpl
import seaborn as sns
import time
import os
sns.set_style('white')
colors = sns.color_palette()
greys  = brewer2mpl.get_map('Greys', 'Sequential', 9)
```

```python
#set parameters
#get the coordinate data from the numerical data   
fp = "withfp"
fpp ="/"+fp
tag = "./1particle"
fname = tag + fpp+"/data.h5"
radius= 8.0
Lx = Ly = 64.0
srate = 0.001
eta = 1.
ss=10
#get  data
h5out = h5py.File(fname,'r')
f0 = h5out["/frame_0/field_data/u_x"].value[:,:,0]
nump, numf = get_h5_pconf(fname)

#prepare arrays.
phi = np.zeros(f0.shape,f0.dtype)
ux = np.zeros(f0.shape,f0.dtype)
uy = np.zeros(f0.shape,f0.dtype)
dx = np.zeros(f0.shape,f0.dtype)
dy = np.zeros(f0.shape,f0.dtype)
r = np.zeros(f0.shape,f0.dtype)
uxdy=np.zeros(f0.shape,f0.dtype)
uydx=np.zeros(f0.shape,f0.dtype)
tauxy=np.zeros(f0.shape,f0.dtype)
u2=np.zeros(f0.shape,f0.dtype)



#particle
ploc = '/frame_'+str(ss)+'/particle_data/'
pos = h5out[ploc+"pos"].value

#coordinate data(or meshgrids avalable)
x = h5out["/coord_data/x"].value[:,:,0]
y = h5out["/coord_data/y"].value[:,:,0]

#get the phi field
count = 0
for i in np.arange(numf/2,numf,1):
    loc = "/frame_"+str(i)+"/field_data/"
    phi = phi + h5out[loc+"phi"].value[:,:,0]
    count = count + 1
phi = phi/float(count)

```

```python
#define the functions which calculate the analytical field

def vel_kernel(ux,x,y,Lx,Ly,a,srate):
    ux[0:Lx-1]=srate*(\
                    -a*a*a*(y[0:Lx-1]-Lx/2.)/(2.*np.sqrt((x[0:Lx-1]-Lx/2.)*(x[0:Lx-1]-Lx/2.)+(y[0:Lx-1]-Lx/2.)*(y[0:Lx-1]-Lx/2.))**(5))\
                    -5.*a*((x[0:Lx-1]-Lx/2.)*(x[0:Lx-1]-Lx/2.)*(y[0:Lx-1]-Lx/2.))\
                    /(2.*np.sqrt((x[0:Lx-1]-Lx/2.)*(x[0:Lx-1]-Lx/2.)+(y[0:Lx-1]-Lx/2.)*(y[0:Lx-1]-Lx/2.))**(5))\
                    *(1-(a*a/(np.sqrt((x[0:Lx-1]-Lx/2.)*(x[0:Lx-1]-Lx/2.)+(y[0:Lx-1]-Lx/2.)*(y[0:Lx-1]-Lx/2.))**(2))))\
                    )
                    

def tau_kernel(sxy,x,y,Lx,Ly,a,srate):
     
      dx = x[0:Lx-1]-Lx/2
      dy = y[0:Ly-1]-Ly/2
      r = np.sqrt(dx[0:Lx-1]*dx[0:Lx-1]+dy[0:Ly-1]*dy[0:Ly-1])
      sxy[0:Lx-1]=srate*(\
                         (-5.*(a**3)/(2.*r*r*r))\
                         +(4.*(a**5)/(r**5))\
                         +((25.*(a**3)*dx*dx*dy*dy)/(r**7))\
                         -((35.*(a**5)*dx*dx*dy*dy)/(r**9))\
                         )
                      
#def tau_test(sxy,x,y,Lx,Ly,a,srate):


#    　sxy[1:Lx-1]=srate*(5.*a*a*(x[1:Lx-1]*y[1:Lx-1])/(2.*(np.sqrt(x[1:Lx-1]*x[1:Lx-1]+y[1:Lx-1]*y[1:Lx-1])**5)))\
#       *(1-(3.*a*a/(np.sqrt(x[1:Lx-1]*x[1:Lx-1]+y[1:Lx-1]*y[1:Lx-1])**2)))
        
def unorm_calc(uxx,uyy,unorm):
    uu = np.dstack((uxx,uyy))
    unorm = np.linalg.norm(uu,axis=2)
```

```python
#define drawer function

def draw_velocity_theory(fig,ax,x,y,skipx,skipy,vscale,width):
        ax.quiver(x[::skipx,::skipy]-Lx/2, y[::skipx,::skipy]-Ly/2,
                   #(1.0-np.ceil(phi[::skipx,::skipy]))*ux[::skipx,::skipy],
                   #(1.0-np.ceil(phi[::skipx,::skipy]))*uy[::skipx,::skipy],
                  (1.0-phi[::skipx,::skipy])*ux[::skipx,::skipy],
                  (1.0-phi[::skipx,::skipy])*uy[::skipx,::skipy],
                  units='xy',angles='uv',scale=vscale,width=width)

def draw_particles(fig, ax, pos, radius):
            i=0
            circ = plt.Circle(pos[i,0:2]-[Lx/2,Ly/2],radius=radius,
                              edgecolor=greys.mpl_colors[8],
                              facecolor='none',lw=2.0)
            ax.add_patch(circ)
            
def draw_reverse_stress(fig, ax, pos, radius):
     scalar = np.sqrt(19./15.)
     for i in range(len(pos)):
            circ = plt.Circle(pos[i,0:2],radius=scalar*radius,
                              alpha=0.3,
                              edgecolor='red',#greys.mpl_colors[8],
                              facecolor='none',lw=2.0)
            ax.add_patch(circ)
            
def draw_stress_theory(fig, ax, phi, sxy, vmin, vmax):
    im = ax.imshow(#sxy.T,
                   ((1.-np.floor(phi))*sxy).T,
                   origin="lower", 
                   interpolation='bicubic', 
                   cmap=bmap.get_mpl_colormap(),alpha=0.5,
                   extent = [-Lx/2,Lx/2-1,-Ly/2,Ly/2-1],
                   vmin=vmin,vmax=vmax)
    
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.15)
    tick =np.arange(vmin,vmax,0.0002)
    cbar= fig.colorbar(im, cax,ticks=tick)
    cbar.ax.tick_params(labelsize=16)

def draw_u2map(fig,ax,unorm,phi,vmin,vmax):
    im=ax.imshow((1.0-np.ceil(phi))*unorm.T, 
                 origin='lower',
                 interpolation="bicubic",
                 cmap=bmap2.get_mpl_colormap(),
                  extent = [-Lx/2,Lx/2-1,-Ly/2,Ly/2-1],
                 alpha=0.6,vmin=vmin, vmax=vmax
                 )
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.15)
    tick =np.arange(0.0,0.0021,0.0005)
    cbar= fig.colorbar(im, cax,ticks=tick)
    cbar.ax.tick_params(labelsize=16)

def draw_streamline(fig,ax,x,y,phi,ux,uy):
     ax.streamplot(y[0]-Ly/2,x[:,0]-Lx/2,(1.0-phi.T)*ux.T,(1.0-phi.T)*uy.T,
                  density=[1.5,1.5],linewidth=1,arrowsize=1.2
                  )
```

```python
#plot the results

bmap = brewer2mpl.get_map('RdBu', 'Diverging',11,reverse=True)
bmap2 = brewer2mpl.get_map('YlOrRd', 'Sequential',8)

fig,   ax = plt.subplots(figsize=(8,7))

ax.xaxis.set_major_locator(LinearLocator(3))
ax.yaxis.set_major_locator(LinearLocator(3))
ax.set_title("analytical",fontsize=18)
ax.set_xlabel('$x$',fontsize = 26)
ax.set_ylabel('$y$',fontsize = 26)
ax.tick_params(labelsize=22)

fig2, ax2 = plt.subplots(figsize=(8,7))
ax2.xaxis.set_major_locator(LinearLocator(3))
ax2.yaxis.set_major_locator(LinearLocator(3))
ax2.set_title("analytical",fontsize=18)
ax2.set_xlabel('$x$',fontsize = 26)
ax2.set_ylabel('$y$',fontsize = 26)
ax2.tick_params(labelsize=22)

skipx=2
skipy=2
a = radius
taurange=0.00040
vscale = 0.00001
width = 0.25

vel_kernel(ux,x,y,Lx,Ly,a,srate)
vel_kernel(uy,y,x,Lx,Ly,a,srate)

tau_kernel(tauxy,x,y,Lx,Ly,a,srate)

tauxy[1:63]=eta*tauxy[1:63]
#tauxy[1:63]=eta*(uxdy[1:63]+uydx[1:63])*a
#unorm_calc(ux,uy,u2)

uu = np.dstack((ux,uy))#make a vector
u2= np.linalg.norm(uu,axis=2)#compute the norm of velocity field

draw_stress_theory(fig,ax,phi,tauxy,-taurange,0.00041)
draw_velocity_theory(fig,ax,x,y,skipx,skipy,vscale,width)        
draw_particles(fig,ax,pos,radius)

u2 =a*a*u2
draw_u2map(fig2,ax2,u2,phi,0,0.00201)
draw_streamline(fig2,ax2,x,y,phi,ux,uy)
draw_particles(fig2,ax2,pos,radius)
#ax.set_xlim(0,Lx)
#ax.set_ylim(0,Ly)
#ax2.set_xlim(0,Lx)
#ax2.set_ylim(0,Ly)
ax.set_xlim(-Lx/2,Lx/2)
ax.set_ylim(-Ly/2,Ly/2)
ax2.set_xlim(-Lx/2,Lx/2)
ax2.set_ylim(-Ly/2,Ly/2)
```
