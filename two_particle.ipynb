```python
%matplotlib inline
from pyrecipes import *
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import brewer2mpl
import seaborn as sns
from IPython.display import Latex,display
from IPython.core.display import Math
from matplotlib.ticker import LinearLocator, FormatStrFormatter
sns.set_style('white')
colors = sns.color_palette()
greys  = brewer2mpl.get_map('Greys', 'Sequential', 9)

gmap = brewer2mpl.get_map('Greys', 'Sequential',6)
bmap = brewer2mpl.get_map('RdBu', 'Diverging',11,reverse=True)
def draw_particles(fig, ax, pos, radius):
    for i in range(len(pos)):
        circ = plt.Circle(pos[i,0:2],radius=radius,
                          edgecolor=greys.mpl_colors[4],
                          facecolor='none',lw=2.0)
        ax.add_patch(circ)
        
def draw_stress(fig, ax, phi, sxy, vmin, vmax):
    im = ax.imshow(((1-phi)*sxy).T, origin='lower', interpolation='bicubic', 
                   cmap=bmap.get_mpl_colormap(),alpha=0.5,
                   vmin=vmin,vmax=vmax)
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.15)
    cbar= fig.colorbar(im, cax)
    tick =np.arange(-0.00040,0.00041,0.0002)
    cbar= fig.colorbar(im, cax,ticks=tick)
    cbar.ax.tick_params(labelsize=16)
    
def draw_vel(fig, ax, x, y, phi, ux, uy, skipx, skipy, vscale):
    ax.quiver(x[::skipx,::skipy], y[::skipx,::skipy],
              (1.0-phi[::skipx,::skipy])*ux[::skipx,::skipy],
              (1.0-phi[::skipx,::skipy])*uy[::skipx,::skipy],
              units='xy',angles='uv',scale=vscale)
```

##About Deviatric stress
In the suspension, the total avarage stress ${<\sigma _{xy}>}$ will be expressed as
$$
<\mathbf{\sigma }>=-<p> \mathbf{I}+2\eta _{f}  \mathbf{<e>}
\underbrace{
+\frac{1}{V}\sum_{i}
\int _{\mathrm{Sur}_{i}} \left(  \mathbf{\sigma}\cdot \mathbf{n}\mathbf{r}-\eta_{f}( \mathbf{u}\mathbf{n}+\mathbf{n}\mathbf{u})\mathrm {d}S
\right)}_{\text{"deviatric stress"}}
$$
,where $\bf{e}$ is the mean shear gradient, ${\mathrm{Sur}_{i}}$ represents the surface of the particle $i$.

Let us call the 3rd term of R.H.S as the "deviatric stress".
The last integral term(without $\frac{1}{V}$ and $\sum _{i}$), say $\mathbf{S}$, corresponds to  the contribution from the existence of the particle.
In other word, $<\mathbf{\sigma }>$ will be changed its value due to the disturbance by the particle.
Therefore, the "deviatric stress" is the total disturbance of the stress by the particle.





##2 particle case
G.K.Batchelor and J. T. Green computed this term  $\mathbf{S}_{ij}$ when the  2 particles moves under the constant shear flow.
They say in that case the $\bf{S}$ would be

$$
\mathbf{S}=\frac{20\pi\eta_{f}a^{3}}{3}\left[
\underbrace{\mathbf{E}\left( 1+K(r) \right)}_{\text{[part 1]}}
+
\underbrace{\left(
\frac{\mathbf {rr\cdot E+E\cdot rr}}{r^{2}}-\frac{\mathbf {r\cdot E\cdot r}}{r^{2}}\frac{2}{3}\mathbf{I}
\right)L(r)
}_{\text{part 2}}
+\\
\underbrace{
\frac{\mathbf{ r\cdot E\cdot r}}{r^{2}}
\left(
\frac{\mathbf {rr}}{r^{2}}-\frac{1}{3}\mathbf{I}
\right)M(r)
}_{\text{[part 3]}}
\right]
$$
Notice that $\mathbf{r}=(x,y,z)$ is the relative coordinate.
I will explain of the parameters.  
If we define $U_{i}$ as the shear flow velocity, the tensor $\mathbf{E}$, which expresses the strain,  is defined as 
$$
\{ \mathbf{E}\} _{ij}= \frac{1}{2}\left[ \frac{\partial U_{i}}{\partial x_{j}}+\frac{\partial U_{j}}{\partial x_{i}}\right]
$$
For the simple DC shear flow, like our simulation, has such the velocity profile that

$$
U_{x} = \dot \gamma y,~~~ U_{y}=U_{z} = 0
$$

Therefore, we have

$$
\mathbf{E} = \frac{\dot \gamma }{2}
\begin{pmatrix}
0&1&0\\
1&0&0\\
0&0&0
\end{pmatrix}
$$

And $K,L,M$ are the nondimensional scalor parameters.
We cannot get the analytical expression, but we can obtain the asymptotic expression when the 2 eqal particles are distant each other.
In that case, we have
$$
K = -\frac{2a^5}{r^5}+\mathcal{O}(\frac{2a^6}{r^{6}})\\
L = -\frac{5a^3}{2r^3} + \frac{10a^5}{r^5}+\frac{25a^6}{4r^6}+\mathcal{O}(\frac{2a^6}{r^{6}})\\
M= \frac{25a^3}{2r^3} -\frac{35a^5}{r^5}+\frac{25a^6}{r^6}+\mathcal{O}(\frac{2a^6}{r^{6}})
$$



##obtain the $S_{xy}$
Here we will calculate the component $S_{xy}$.
Let us separate the $\mathbf{S}$ into 3 parts, which contains $(1+K),L,M$.
The $xy$ component of the first term obiously 

$$\text{[part 1] }=E_{xy}(1+K)=(\dot \gamma /2)(1+K)$$

About second component, notice that $\delta _{xy}=0$, so 

$$E_{kl}\cdot \frac{r_x r_k \delta _{yl}+r_y r_k \delta _{xl}}{r^2}$$

Only the 2 components $E_{xy},E_{yx}$ are non zero, therefore,

$$\text{[part 2]}=\frac{\dot \gamma}{2}\frac{x^2 +y^2}{r^2}$$

As the same way as part 2, we find that

$$\text{[part 3]}=E_{kl} \frac{r_{k}r_{l}}{r^2}=\frac{\dot \gamma}{2}\frac{2x^2 y^2}{r^4}$$

At a result, we obtain

$$
S_{xy}=\frac{20\pi\eta_{f}a^{3}}{3}\frac{\dot \gamma}{2}
\left[
(1+K)+\frac{x^2 +y^2}{r^2}L+\frac{2x^2 y^2}{r^4}M
\right]
$$

the deviatric stress can be obtained from the equation below;

$$
<\sigma> _{xy}=2\eta _{f}  \mathbf{<e>}_{xy}+\frac{2}{V}S_{xy}
$$
,so $\frac{2}{V}S_{xy}$  is the deviatric stress(notice that the $\sum_{i}$ yields 2, number of particles)
