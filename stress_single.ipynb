```python
%matplotlib inline
from pyrecipes import *
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import brewer2mpl
import seaborn as sns
sns.set_style('white')
colors = sns.color_palette()
greys  = brewer2mpl.get_map('Greys', 'Sequential', 9)
```

```python
#tag = "phi0.001_s0.004/"
fp = "/withoutfp"
tag = "./1particle"
fname = tag + fp+"/data.h5"
radius= 8.0
Lx = Ly = 64.0
srate = 0.001
```

```python
nump, numf = get_h5_pconf(fname)#nump particle number, numf number of frames
print nump
print numf
```

```python
h5out = h5py.File(fname,'r')
x = h5out["/coord_data/x"].value[:,:,0]
y = h5out["/coord_data/y"].value[:,:,0]
def draw_particles(fig, ax, pos, radius):
    for i in range(len(pos)):
        circ = plt.Circle(pos[i,0:2],radius=radius,
                          edgecolor=greys.mpl_colors[4],
                          facecolor='none',lw=2.0)
        ax.add_patch(circ)

for i in np.arange(50,51,1):
    ploc = '/frame_'+str(i)+'/particle_data/'
    pos = h5out[ploc+"pos"].value
    fig, axt = plt.subplots(figsize=(7,7))
    draw_particles(fig,axt,pos,radius)
print x.shape
print y.shape
```

```python
count = 0

f0 = h5out["/frame_0/field_data/u_x"].value[:,:,0]
ux = np.zeros(f0.shape, f0.dtype)
uy = np.zeros(f0.shape, f0.dtype)
tauxy = np.zeros(f0.shape,f0.dtype)
phi = np.zeros(f0.shape,f0.dtype)
us = srate*(y-Ly/2.0)



for i in np.arange(numf/4,numf,1):
    loc = "/frame_"+str(i)+"/field_data/"
    ux = ux + h5out[loc+"u_x"].value[:,:,0]
    uy = uy + h5out[loc+"u_y"].value[:,:,0]
    tauxy = tauxy + h5out[loc+"tau_xy"].value[:,:,0]
    #phi = phi + h5out[loc+"phi"].value[:,:,0]
    count = count + 1
    phi = phi + h5out[loc+"phi"].value[:,:,0]
ux = ux / float(count)
uy = uy / float(count)
tauxy = tauxy/float(count)
phi = phi/float(count)

uxx = ux-us
uu = np.dstack((uxx,uy))
u2 = np.linalg.norm(uu,axis=2)
```

```python
sx = sy = 3;
fig, [ax,ax2,ax3]= plt.subplots(3,1,sharex=True, figsize=(5,10))
#bmap = brewer2mpl.get_map('YlGnBu', 'Sequential',8)
bmap = brewer2mpl.get_map('RdBu', 'Diverging',8)
bmap2 = brewer2mpl.get_map('Greys', 'Sequential',3)

#ax.quiver(x[::sx,::sy],y[::sx,::sy],uxx[::sx,::sy],uy[::sx,::sy],
#          units='xy',angles='uv',scale=srate*radius*0.003)
ax.streamplot(y[0],x[:,0],uxx.T,uy.T, density=[1.3,1.2], linewidth=1, arrowsize=1.2,
              color=greys.mpl_colors[6])

im=ax.imshow(u2.T/srate, origin='lower', interpolation="bicubic",
             cmap=bmap.get_mpl_colormap(),alpha=0.6,vmin=0, vmax=0.1*10)
im2=ax2.imshow(tauxy.T/srate, origin='lower', interpolation="bicubic",
             cmap=bmap.get_mpl_colormap(),alpha=0.6,vmin=-0.5, vmax=0.5)
im3=ax3.imshow(phi, origin='lower', interpolation="bicubic",
             cmap=bmap2.get_mpl_colormap(),alpha=0.6)#,vmin=-0.5, vmax=0.5)
divider = make_axes_locatable(ax)
divider2 = make_axes_locatable(ax2)
divider3 = make_axes_locatable(ax3)
cax = divider.append_axes("right", size="5%", pad=0.15)
cax2 = divider2.append_axes("right", size="5%", pad=0.15)
cax3 = divider3.append_axes("right", size="5%", pad=0.15)
cbar= fig.colorbar(im, cax)
cbar2= fig.colorbar(im2, cax2)
cbar3= fig.colorbar(im3, cax3)
ax.set_xlim(0,Lx)
ax.set_ylim(0,Ly)
ax2.set_xlim(0,Lx)
ax2.set_ylim(0,Ly)
ax3.set_xlim(0,Lx)
ax3.set_ylim(0,Ly)
#fig.savefig("single_u.pdf")
```
